```python
#imports
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

from geopy.geocoders import Nominatim 

import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

df = pd.read_csv('C:\\xampp\\htdocs\\tourism\\tourism\\customer.csv', sep=',')
```

```python
#geopy in action
country = 'India'
city_names = df['td']

longitude =[]
latitude =[]
geolocator = Nominatim(user_agent="Trips", timeout= 10)
#under user_agent any thing works here mentioned as Trips

for c in city_names.values:
    location = geolocator.geocode(c+','+ country)
    latitude.append(location.latitude)
    longitude.append(location.longitude)
```

```python
df['Latitude']=latitude
df['Longitude']=longitude
```

```python
l2=df.iloc[:,-1:-3:-1]
l2
```

```python
kmeans=KMeans(5)
kmeans.fit(l2)
identified_clusters=kmeans.fit_predict(l2)
identified_clusters=list(identified_clusters)
identified_clusters
```

```python
df['loc_clusters']=identified_clusters
df
```

```python
input_city = input("Enter a city name: ")
cluster = df.loc[df['td'] == input_city, 'loc_clusters']
cluster = cluster.iloc[0]
cluster
cities = df.loc[df['loc_clusters'] == cluster, 'td']
cities
for c in range(len(cities)):
    if cities.iloc[c] == input_city:
        continue
    else:
        print(cities.iloc[c])
```
